```python
#Q1
import yfinance as fac
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
aapl=fac.download("AAPL",period="6mo")
msft = fac.download("MSFT",period="6mo")
google= fac.download("GOOGL" , period="6mo")
aaplclose = aapl['Close','AAPL'].tolist()
msftclose= msft['Close','MSFT'].tolist()
googleclose = google['Close','GOOGL'].tolist()
data = [aaplclose,msftclose,googleclose]
df= pd.DataFrame(data,index =['aapl','msft','google'])
df.columns = aapl.index
dft = df.T
display(dft)
investment = 10000/3
dailyreturn = dft.pct_change()
print(dailyreturn)
nofstocks =3
w = [1/nofstocks]*nofstocks
print(w)
portfoliodailyreturn= dailyreturn.dot(w)
print(portfoliodailyreturn)
plt.plot(portfoliodailyreturn)
plt.xlabel("Date")
plt.ylabel("Daily Portfolio Return")
plt.title("dpr vs date")
```

```python
#Q3 
import yfinance as fac
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
n=5
t =[]
w=[]
for i in range (n):
    tickers = input("Enter a ticker: ")
    t.append(tickers)
    weights = float(input("Enter a weight: "))
    w.append(weights)

sdi = input("Enter start date(YYYY-MM-DD): ")
startdate=datetime.strptime(sdi,"%Y-%m-%d").date()
edi = input("Enter end date(YYYY-MM-DD): ")
enddate = datetime.strptime(edi,"%Y-%m-%d").date()

t1 = fac.download(t[0], start = startdate, end = enddate)['Close']
w1 = w[0]
dt1 = t1.pct_change()

t2=fac.download(t[1], start = startdate, end = enddate)['Close']
w2 = w[1]
dt2 = t2.pct_change()

t3 = fac.download(t[2], start = startdate, end = enddate)['Close']
w3 = w[2]
dt3 = t3.pct_change()

t4 = fac.download(t[3], start = startdate, end = enddate)['Close']
w4 = w[3]
dt4 = t4.pct_change()

t5 = fac.download(t[4], start = startdate, end = enddate)['Close']
w5 = w[4]
dt5= t5.pct_change()

dt1list = dt1[t[0]].tolist()
dt2list = dt2[t[1]].tolist()
dt3list = dt3[t[2]].tolist()
dt4list = dt4[t[3]].tolist()
dt5list = dt5[t[4]].tolist()
data = [dt1list,dt2list,dt3list,dt4list,dt5list]
df = pd.DataFrame(data, index=t)
df.columns = t1.index
dft = df.T

portfoliodailyreturn = dft.dot(w)
print(portfoliodailyreturn)

plt.plot(portfoliodailyreturn)
plt.xlabel('Date')
plt.ylabel('dpr')
plt.title('dpr vs date')


```
